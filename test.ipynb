```python
import matplotlib.pyplot as plt
import matplotlib as mpl

import numpy as np
```

```python
button_cmap = mpl.colormaps['autumn']
screen_cmap = mpl.colormaps['winter']

num_buttons = 2
num_screens = 10

bc = button_cmap(np.linspace(0, 1, num_buttons))
sc = screen_cmap(np.linspace(0, 1, num_screens))

fig, ax = plt.subplots(layout='constrained')

for i, color in enumerate(bc):
    ax.plot([0, i], color=color)

plt.show()
```
